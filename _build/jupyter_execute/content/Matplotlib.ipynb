# ðŸ“Š Matplotlib for Data Visualization
Matplotlib is a widely used Python library for creating static, animated, and interactive plots.

## ðŸ“¥ Importing Libraries

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

## ðŸ“ˆ Line Plot

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y)
plt.title('Sine Wave')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.grid(True)
plt.show()
```

## ðŸ“‰ Multiple Lines on One Plot

```python
y2 = np.cos(x)
plt.plot(x, y, label='sin(x)')
plt.plot(x, y2, label='cos(x)')
plt.legend()
plt.title('Sine and Cosine')
plt.show()
```

## ðŸ“Š Bar Plot

```python
labels = ['A', 'B', 'C']
values = [10, 24, 36]
plt.bar(labels, values)
plt.title('Bar Plot')
plt.ylabel('Value')
plt.show()
```

## ðŸ“¦ Histogram

```python
data = np.random.randn(1000)
plt.hist(data, bins=30, color='skyblue', edgecolor='black')
plt.title('Histogram')
plt.show()
```

## ðŸ“ Scatter Plot

```python
x = np.random.rand(50)
y = np.random.rand(50)
colors = np.random.rand(50)
sizes = 1000 * np.random.rand(50)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5)
plt.title('Random Scatter Plot')
plt.show()
```

## ðŸ–¼ï¸ Subplots

```python
# Dummy data
x = np.arange(10)
y = np.sin(x)
y2 = np.cos(x)  # same size as x
labels = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J']
values = np.random.randint(1, 10, size=10)
data = np.random.randn(100)

# Create 2x2 grid of plots
fig, axs = plt.subplots(2, 2, figsize=(10, 6))

axs[0, 0].plot(x, y)
axs[0, 1].bar(labels, values)
axs[1, 0].hist(data, bins=20)
axs[1, 1].scatter(x, y2)

plt.tight_layout()
plt.show()
```
