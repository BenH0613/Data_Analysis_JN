## Quick Introduction to Common matplotlib Syntax

Here are the most useful commands youâ€™ll use in matplotlib:

### Creating Basic Plots

```python
plt.plot(x, y)            # Line plot
plt.bar(x, heights)       # Bar chart
plt.scatter(x, y)         # Scatter plot
```
### Labels and Titles
```python
plt.xlabel("X-axis label")       # Label for the x-axis
plt.ylabel("Y-axis label")       # Label for the y-axis
plt.title("My Plot Title")       # Add a title

```
### Style and appearance
```python
plt.plot(x, y, marker='o')       # Add markers at data points
plt.plot(x, y, color='green')    # Change line color
plt.grid(True)                   # Show grid lines
plt.legend()                     # Show a legend (with label= in plot)
plt.figure(figsize=(8, 5))       # Set the size of the plot (width, height)

```
### Saving the plot
```python
plt.savefig("myplot.png")        # Save the plot as an image file



## Importing Libraries

````python
# Import matplotlib
import matplotlib.pyplot as plt
import numpy as np

# Enable inline plotting (for Jupyter Notebooks)
%matplotlib inline
````

## Line Plot

````python
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y)
plt.title('Sine Wave')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.grid(True)
plt.show()
````

##  Multiple Lines on One Plot

````python
y2 = np.cos(x)
plt.plot(x, y, label='sin(x)')
plt.plot(x, y2, label='cos(x)')
plt.legend()
plt.title('Sine and Cosine')
plt.show()
````

## Bar Plot

````python
labels = ['A', 'B', 'C']
values = [10, 24, 36]
plt.bar(labels, values)
plt.title('Bar Plot')
plt.ylabel('Value')
plt.show()
````

## Histogram

````python
data = np.random.randn(1000)
plt.hist(data, bins=30, color='skyblue', edgecolor='black')
plt.title('Histogram')
plt.show()
````

## Scatter Plot

````python
x = np.random.rand(50)
y = np.random.rand(50)
colors = np.random.rand(50)
sizes = 1000 * np.random.rand(50)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5)
plt.title('Random Scatter Plot')
plt.show()
````

## Subplots

````python
# Dummy data
x = np.arange(10)
y = np.sin(x)
y2 = np.cos(x)  # same size as x
labels = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J']
values = np.random.randint(1, 10, size=10)
data = np.random.randn(100)

# Create 2x2 grid of plots
fig, axs = plt.subplots(2, 2, figsize=(10, 6))

axs[0, 0].plot(x, y)
axs[0, 1].bar(labels, values)
axs[1, 0].hist(data, bins=20)
axs[1, 1].scatter(x, y2)

plt.tight_layout()
plt.show()
````

## Example: Enzyme Activity Over Time

You measured the activity of an enzyme every 5 minutes.

Letâ€™s plot this data as a line chart to visualize how activity increases over time.

````python
# Experimental data
time_min = [0, 5, 10, 15, 20, 25, 30]
activity_umol_min = [0.0, 1.9, 3.8, 5.3, 6.8, 7.5, 8.0]

# Create a line plot
plt.plot(time_min, activity_umol_min)

# Add labels and title
plt.xlabel("Time (minutes)")
plt.ylabel("Enzyme Activity (Âµmol/min)")
plt.title("Enzyme Activity Over Time")

# Show the plot
plt.show()
````

## Customizing Your Plot

Letâ€™s make the plot more readable by:

- Adding markers
- Changing the line style
- Making the plot size larger

````python
# Create a more customized plot
plt.figure(figsize=(8, 5))  # Set plot size

plt.plot(time_min, activity_umol_min,
         marker='o',         # Add circular markers
         linestyle='-',      # Solid line
         color='green',      # Line color
         label='Activity')   # Legend label

# Add title and axis labels
plt.title("Enzyme Activity Over Time", fontsize=14)
plt.xlabel("Time (minutes)")
plt.ylabel("Activity (Âµmol/min)")

# Show legend
plt.legend()

# Show grid
plt.grid(True)

# Show plot
plt.show()
````

## Bar Plot: Comparing Treatment Groups

Letâ€™s say you ran 3 treatments in an ELISA assay and want to compare protein concentrations.

Bar plots are great for comparing values across categories.

````python
# Sample data
groups = ['Control', 'Treated A', 'Treated B']
means = [1.2, 2.6, 2.1]

# Create a bar plot
plt.bar(groups, means, color=['blue', 'orange', 'green'])

# Add labels and title
plt.title("Average Protein Concentration")
plt.ylabel("Protein (Âµg/mL)")

# Show plot
plt.show()
````

## Saving Your Plot as an Image

You can save any plot using `plt.savefig()` to use in your reports or presentations.

````python
# Save the enzyme activity plot
plt.figure()
plt.plot(time_min, activity_umol_min, marker='o')
plt.xlabel("Time (minutes)")
plt.ylabel("Activity (Âµmol/min)")
plt.title("Enzyme Activity")
plt.savefig("enzyme_activity_plot.png")  # Saves to current directory
plt.show()
````
